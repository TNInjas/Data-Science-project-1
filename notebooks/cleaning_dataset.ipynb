```python
# Goal : Cleaning the dataset and return a final clean dataset.
import pandas as pd
```

```python
df = pd.read_csv("financial_transaction_test.csv")
df
```

```python
df["Transaction_Text"][1]
```

```python
df.loc[0, "Transaction_Text"]
```

```python
price_list = []
for i in range(0, 999 + 1):
    k = (df["Transaction_Text"].str.split(r'^[^|]*\|[^|]*\|\s* Amount: INR'))[i][1].replace(" ", "")
    price_list.append(float(k))
```

```python
r = {
    "Categories": df["Label"],
    "Amount spent": price_list
}
```

```python
f = pd.DataFrame(r)
f
```

```python
f.to_csv("cleaned_dataset.csv", index=False)
```
